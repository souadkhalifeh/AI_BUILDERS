```python
import pandas as pd
```

```python
car_sales = pd.read_csv("../pandas/car-sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[$,.]', '', regex=True)
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sales Date"] = pd.date_range("1/1/2026", periods= len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x='Sales Date', y = 'Total Sales')
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.plot(x= "Odometer (KM)", y ="Price", kind= "scatter")
```

```python
car_sales["Price"] = car_sales["Price"].astype(str)
car_sales.plot(x= "Odometer (KM)", y ="Price", kind= "scatter")
```

```python
car_sales.plot(x='Make', 
               y = 'Odometer (KM)', 
               kind = 'bar')
```
